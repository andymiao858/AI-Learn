##二项分布
设试验$E$只有两个可能结果:$A$及$\overline{A}$,则称$E$为二项分布，又叫做伯努利（Bernoulli）试验。
设$P(A)=p \quad (0<p<1)$。此时$P(\overline{A})=1ーp$。
将E独立重复地进行n次，则称这串重复的独立试验为n重伯努利试验。

<br/>

那现实中什么事件符合二项分布呢，最典型的就是扔硬币。
假设扔10次硬币，4次朝上的概率是多少？
假设投掷一次，硬币朝上的概率是$p$，则硬币朝下的概率就是$1-p$。
实际中，我们得知，$p$一般为0.5。
则N次朝上的概率为
<br/>

$P(N=4) = C_{10}^{4}p^4(1-p)^6=0.205$

<br/>
根据经验，我们得知现实中，我们抛10次硬币，朝上的次数分别是3，4，5，6，7的概率应该是比较大的，那我们就算算到底概率是多大
<br/>

$P(N=3,4,5,6,7) = C_{10}^{3}p^3(1-p)^7 + C_{10}^{4}p^4(1-p)^6 + C_{10}^{5}p^5(1-p)^6 + C_{10}^{6}p^6(1-p)^4 + C_{10}^{7}p^7(1-p)^3=0.891$


那知道这个概率有什么用呢？
如果有庄家开了一个赌局。抛10次硬币，如果3，4，5，6，7次朝上赔率是1.1，其他情况赔率是9。
假设可以无限投，你该如何投呢？
假设投100元，朝上，按照上面的概率计算，我得到的钱的期望是多少呢？
<br/>
$E=0.891*100*1.1-0.109*0=98.01$

我得到的钱的期望比我的本金100元还要少？这不是赔本的买卖吗，精通概率计算的我一定不会投朝上。

<br/>
那么，我来赌把大的吧！如果小概率发生，可是9倍的赔率啊，100变900啊，碰碰运气吧，慢。。。
先停下来算算概率。
<br/>
$E=0.109*100*9-0.891*0=98.1$

还是比我的本金少？又是赔本的买卖，庄家真是黑啊，投哪个都是赔。
所以**珍爱生命，学好概率，远离赌博。**

那么上面庄家的赔率是怎么算出来的呢？其实很简单。
如果想让投3，4，5，6，7朝上的人不赚不赔，那么得到钱的期望一定就等于本金

$E * 0.891 = 100$,
$E = 100/0.891 = 112$
赔率 = 112/100 = 1.12

所以，庄家如果想让你赔钱，只要把赔率稍微调小一点点，变成1.1，在多次赌博的情况下，你就是一定是赔钱的了。

另外一方呢，赔率=1/0.109=9.17，庄家又把赔率调小一点点，变成9，也是赔钱的。

<br/>
那么谁是赚钱的呢？答案当然是庄家。假设有100个人参与赌博，那么资金池里就有100*100=1w元。
有N个人投3，4，5，6，7朝上，100-N个人投另外一方，那么他们分走钱的期望是
$E=0.891*110*N + 0.109*900*(100-N) = 9810-0.09*N <= 9810 $

所以资金池里剩下了190就被庄家拿走了，庄家啥也不干，稳赚190，所以要当好庄家，必须学好概率论。

## 泊松分布
泊松分布适合于描述单位时间内随机事件发生的次数的概率分布。

日常生活中，大量事件是有固定频率的。这种事件我们可以用泊松分布来建模。如某一服务设施在一定时间内受到的服务请求的次数、机器出现的故障数、某个超市每天卖出的饮料个数、某个妇产医院每天新生的婴儿个数等等。

这类事件的特点是，我们可以求出事件平均发生率，比如某妇产医院平均每天出生50个婴儿。但我们无法知晓今天出生多少婴儿？

泊松分布的概率质量函数为：

$P(X=k)= \frac{e^{-\lambda}\lambda^k}{k!} $

泊松分布的参数 $\lambda$ 是单位时间（或单位面积）内随机事件的平均发生率。

假设我是一个卖煎饼果子的，那我应该每天准备多少煎饼果子才能利润最大化呢？也就是人家想买煎饼果子我不至于卖完了，另外还不能剩太多，否则剩下的材料太多也容易赔。
于是，我开始记录周一到周五卖煎饼个数的数据
周一：10， 周二：15，周三：8，周四13，周五：18
平均下来，我每天大概能卖

$\frac{10+15+8+13+18}{5}=12.8$

所以我们可以假设平每天卖煎饼果子个数大概在13，我们可以认为13是我们每天卖煎饼果子的平均个数，则泊松分布如下

$P(X=k)= \frac{e^{-13}13^k}{k!} $

如果我每天按照平均数进行准备，那我原材料到底够不够用呢？
我们可以用泊松分布计算一下

程序如下：


```
import math

COEF = math.exp(-13.0)

def f(n):
  if n == 1:
    return 1
  else:
    return n * f(n - 1)

sum = 0

for n in range(1, 13):
  sum += COEF*(13**n)/ f(n)

print("Sum: %f" % sum)     
```

通过程序，我们可以求出，只有46%的可能性原材料是足够的，也就是一周，有2-3天是不够卖的，这样卖煎饼果子，有钱不赚可是不行的。

那么，我到底应该准备多少原材料呢？理论上，准备的越多，肯定越够卖。如果我每天准备100个，那卖不掉剩下的煎饼果子把我卖出去的利润都吃光了也不行啊。


<br/>
如果我想在95%的情况下都能保证原料足够，我应该每天准备多少呢？

$P(X \leq k)=\sum_{i=1}^{k}\frac{e^{-13}13^k}{k!} \geq\ 0.95$

程序如下：

```
import math

COEF = math.exp(-13.0)

def f(n):
  if n == 1:
    return 1
  else:
    return n * f(n - 1)

sum = 0
n = 1

while(True):
  sum += COEF*(13**n)/ f(n)
  if (sum >= 0.95):
    break
  else:
    n = n + 1   

print("Sum: %f" % sum)
print("Iteration: %d" % n)      
```

通过计算求得，当我们准备19个煎饼果子的时候，能保证95%以上原材料是足够的。对于剩下的5%的小概率事件，我们暂时忽略了，即使哪天万一不够了，就早点放假回家休息了。
所以，卖煎饼果子也需要概率知识啊。

## 正态分布
正态分布是我们生活中非常常见。如果把一个学校男生的身高统计一下，横轴是身高，纵轴是人数，则它的形状会和正态分布函数非常像，也就是我们经常说的钟形曲线，中间高，两侧低。
再比如体重等都符合正态分布。

为什么现实世界中正态分布如此之多？
说这个之前，先介绍一下最大熵原理。
这里面的**熵**在热力学中是物体的混乱程度，在信息论中，用来描述信息的不确定程度。

在热力学中，熵增原理就是孤立热力学系统的熵不减少，总是增大或者不变。用来给出一个孤立系统的演化方向。说明一个孤立系统不可能朝低熵的状态发展即不会变得有序。

比如，在原来有半杯凉水的杯中倒入半杯热水，很长时间以后，杯里的水一定是冷热均匀的。
而一杯温水，如果放在那里不动，是无论如何也无法变成上面热，下面冷的。
因为上热下冷是有序的，也就是混乱程度低，也就是熵低。一个孤立的系统，只可能朝熵增的方向变化，也就是冷热水混在一起。

<br/>
在信息论中，熵用来描述信息的不确定程度，其实也就是信息的混乱程度。最大熵原理就是，在给定的约束条件下，满足最大熵的时候的将信息的不确定性保留到最大，也就是在作出了最合理的估计，在估计出错时候的代价最小。
比如，掷一个骰子，在不知道任何信息的时候，1-6的概率都是$\frac{1}{6}$。在不知不觉中，我们就用到了最大熵原理，也就是我们在不知道任何信息的时候，等概率是最好的假设。什么是最好的假设，就是代价函数最小。比如我掷骰子赌博，在1-6投相同的钱我损失的钱是最少的。否则，我在1上投更多的钱，最后的结果一定是损失比平均投要损失的更多（当然，假设骰子是均匀的）。


从上面**熵**的描述中，可以大概体会到熵的感觉。
宇宙一定是越来越混乱，最后沉寂。大自然如果不做干涉，一定是最终以熵最大的形式存在。

<br/>

现实生活中，各种事件的分布都是未知的，但一般均值$\mu$和方差$\sigma$是固定的，那在这两个值确定的前提下，什么分布是熵最大的呢？答案就是**正态分布**。大自然是一个非常高超的数学家和设计师，精美的设计了万物，最大熵原理竟然和正态分布联系在一起。


若随机变量$X$服从一个位置参数为$\mu$ 、尺度参数为${\sigma }$ 的正态分布，记为：
${X \sim N(\mu,\sigma^2)}$
则其概率密度函数为

${f(x)={\frac{1}{\sigma {\sqrt {2\pi }}}}\,e^{{-{(x-\mu )^{2}} \over {2\sigma ^{2}}}}}$

正态分布的数学期望值或期望值$\mu$ 等于位置参数，决定了分布的位置；其方差$\sigma^2$的开平方或标准差$\sigma$ 等于尺度参数，决定了分布的幅度。


当然，从数学上**中心极限定理**也说明了为什么很多分布符合正态分布。

*特定条件下，大量统计独立的随机变量的平均值的分布趋于正态分布，这就是**中心极限定理**。*

<br/>
