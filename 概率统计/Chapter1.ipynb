# 概率与统计
## 概率论

世界上不是所有的现象都是因果关系，比如朝天上扔了一个石头，它一定会掉下来。
在计算器上输入1+1， 一定会输出2。

现实中，很多现象是以概率存在的。比如

<br/>
明天下雨的概率是多大？

我们无法获得电子在原子的运动轨迹，也就是测不准原理，那我们能不能获得大量电子运动的统计规律？

<br/>
概率论作为一门数学分支，它所研究的内容一般包括随机事件的概率、统计独立性和更深层次上的规律性。
概率论一般包括**古典概率论**和**现代概率论**等。

<br/>
比如：

*一个盒子里有4个红球，6个黑球，我从盒子里抽出一个红球的可能是多大？*

*我买了一个35选7的彩票，得一等奖的可能是多大？*

<br/>
研究这类问题，一般是**古典概率论**研究的东西。
当然还有现代概率论，它研究的是随机过程。

<br/>
可能感觉到了，古典概率论（又叫拉普拉斯概率）研究的是有限事件的问题，而且每个事件发生的可能性相等。
现实世界中，每个事件发生的概率不一定完全相等。比如，我掷骰子，骰子的分布不一定均匀，也就是得到1，2，3，4，5，6每个数的概率不一定如我们期望，是均等的。但我们不知道骰子到底是否不均匀，所以我们只能假设每一面出现的概率相等。
那么**每个事件发生的可能性相等**背后的理论依据是什么呢？其实是**最大熵原理**。这个在后面会介绍，现在需要知道的就是，在没有足够已知条件时，选取各种可能性概率相等是一种有效的处理准则。这个从数学上是可以证明的。

有了古典概率论，现代概率论研究的就是事件不定的，也就是随机事件，我们无法用有限的事件来衡量，它的概率只能是一种分布函数，而如何构建分布函数，分布函数的参数是什么，我们甚至无法精确的测量。比如股票，汇率的波动。我们预测明天股票涨的可能是多大，这里面有太多的影响因子，每个影响因子的权重是多大等等？当然这就是一个例子，现实中，股票的涨跌受太多因素影响了，根本无法找到一个合适的模型去预测。以上就是现代概率论要研究的东西。

所以，概率论是已知了各种条件和参数，去研究某件事情发生的概率。这有点像演绎推理法，知道了原理和原则，一步一步推导，就能推导出正确的结果。

## 统计
在概率论中,我们所研究的随机变量，它的分布都是假设已知的,在这一前提下去研究它的性质、特点和规律性，例如求出它的数字特征，讨论随机变量函数的分布,介绍常用的各种分布等。
在数理统计中，我们研究的随机变量，它的分布是未知的，或者是不完全知道的，人们是通过对所研究的随机变量进行重复独立的观察，得到许多观察值，对这些数据进行分析，从而对所研究的随机变量的分布作出种种推断的

也就是统计是基于数据来反推模型和参数。

<br/>
比如，

*我想估计一下学校里男生，女生的比例。于是，我蹲在门口，看校门出入的男女生比例，校门出入的一共1000个人，700个男生，300个女生。我可能会估计，这个学校男生，女生的比例大概是7:3。于是根据这个估算，我可以预估后来经过校门口的人70%可能是男生，30%可能是女生。*

*但是如果样本减小到10个，经过校门的一共10个人，其中10个都是男生，我能说学校里都是男生吗？当然不能。*

<br/>
所以，其实在做统计的时候，对数据量是有要求的，过小的数量预估出来的模型是不准确的。数据量和可信度之间的关系，我们后面再说。

可以看出来，统计有点像的“归纳法”。我不知道模型是什么，于是给我一堆数据，我从中归纳出模型的参数和特性。当我得知了模型和参数以后，我就可以利用它去做预测，也就是前面“概率论”研究的东西。

在AI中，统计对应的就是训练过程，概率研究的就是前向计算的问题。
所以，概率与统计是研究问题的两个方面，缺一不可。
