# 协方差

现实世界中有很多事物的联系不是因果关系，而是相关关系。
比如

<br/>

**啤酒与尿布**

*我们去沃尔玛超市会发现一个很有趣的现象：货架上啤酒与尿布竟然放在一起售卖，这看似两者毫不相关的东西，为什么会放在一起售卖呢？*

*原来，在美国，妇女们经常会嘱咐她们的丈夫下班以后给孩子买一点尿布回来，而丈夫在买完尿布后，大都会顺手买回一瓶自己爱喝的啤酒（由此看出美国人爱喝酒）。商家通过对一年多的原始交易记录进行详细的分析，发现了这对神奇的组合。于是就毫不犹豫地将尿布与啤酒摆放在一起售卖，通过它们的关联性，互相促进销售。*

我们不能说买啤酒一定会买尿布，但是买啤酒和买尿布之间会有一定的正向关系。  

<br/>

**美元与黄金**

*如果我们观察美元汇率和黄金的走势，经常会发现他们成相反的走势。因为如果美元汇率走强，大家会购买美元，而减少黄金的投资，从而黄金的价格会下跌。同时，以美元为标的的黄金价格也会下跌。当美元走弱时，大家会觉得美元不能保值，于是大量的购买黄金以规避风险，起到保值的作用。*

所以，历史上来看，美元和黄金在绝大部分时候是走势相反的。但是我们不能看美元跌，立即购入黄金，因为他们没有因果关系，而且关联关系到底多大，我们也无从知晓。

<br/>

刚才说到了“正向关系”，“走势相反”，从专业术语上来说，其实就是相关性。

前一种情况是正相关，后一种情况是负相关。无论是正相关还是负相关，如果相关度足够高，我们都可以根据其中一个事件的走势大概率的预测出另一个事件的走势。


这里面的相关性，用专业术语来说就是**“协方差”**。

$Cov(X, Y) = E\big \{{[X - E(X)][Y - E(Y)]}\big \} $

<br/>

其实就是比较 $X$，$Y$ 跟它的期望值 $E(X)$, $E(Y)$ 的差，如果$X, Y$的变化趋势相同，则$Cov$的值会是一个很大的正数。如果$X, Y$的变化趋势相反，则$Cov$的值会是一个很小的负数。
无论很大或者很小，我们都认为$X, Y$是相关性很高的，只不过是正相关还是负相关而已。

当然，如果$Cov(X, Y) = 0$，我们可以认为$X, Y$没有任何相关性。

<br/>

实际使用中，由于$X, Y$两个变量的单位可能不同，比如一组变量的变化范围是0-1，而另一组变量的变化范围是0-10000。那么即使第二组的相关性低，但是由于数值变化范围大，导致协方差会更大。所以导致协方差没有可比性，协方差大的不一定比协方差小的相关性更高。

于是，我们定义了**相关系数**

<br/>

$\rho_{XY} = \frac{Cov(X, Y)}{\sqrt{D(X)} \sqrt{D(Y)}}$
其中$D$表示方差。

我们可以认为方差是衡量变量变化幅度的一个函数，这样我们就把$Cov$归一化了。数学上可以证明，相关系数$\rho_{XY}$的变化范围是$[-1,1]$。

<br/>
这样，我们就可以就可以通过相关系数定量的分析两个变量之间的相关性。
